```python
resnext = torch.load('../pretrained/resnext.pt')
resnext = torch.nn.Sequential(*(list(resnext.children())[:-1]))
resnext.cuda()
```

```python
# Feature Extractor
def transform(d, model):
    data = []
    for idx, di in enumerate(d):
        img = cv2.imread(di[0])
        img = cv2.resize(img, (700,460))
        img = np.reshape(img, (3,460,700))
        img = torch.tensor(np.expand_dims(img, axis=0))
        img = img.type(torch.FloatTensor)
        img = img.cuda()
        img = model(img).cpu().detach().numpy().reshape((-1, 1))
        data.append([img, di[1]])
    data = np.array(data)
    return data
```
